### Preprocessing Brown corpus 

```python
sentence = brown.sents()
extrawords=[',','.',"''",'``']
```

```python
unigram=[]
bigram=[]
trigram=[]
fivegram=[]
tengram=[]
```

### Training the n gram models

```python
for s in sentence:
    s = list(map(lambda x:x.lower(),s))
    for w in s:
        if w in extrawords:
            s.remove(w) 
        else:
            unigram.append(w)
    bigram.extend(list(ngrams(s, 2,pad_left=True, pad_right=True)))
    trigram.extend(list(ngrams(s, 3, pad_left=True, pad_right=True)))
    fivegram.extend(list(ngrams(s, 5, pad_left=True, pad_right=True)))
    tengram.extend(list(ngrams(s, 10, pad_left=True, pad_right=True)))
```

```python
f1 = FreqDist(unigram)
f2 = FreqDist(bigram)
f3 = FreqDist(trigram)
f5 = FreqDist(fivegram)
f10 = FreqDist(tengram)
```


### Downloading and Preprocessing APPLING1DAT.643



```python
with open('APPLING1DAT.643') as f:
    lines=f.readlines()
```

```python
cword=[]
incword=[]
test=[]
i=1
for l in lines:
    if l.startswith('$'):
        continue
    else:
        l.replace('\n','')
        w = l.split()
        cword.append(w[1])
        incword.append(w[0])
        test.append(w[2:])
```
